```python
class YourClass:
    
    def say_hello():
        print('Hello')
        
    def instance_say_hello(self):
        print('hello {}'.format(self))
        #print(f'Hello from{self}')
        
    @classmethod
    
    def class_say_hello(cls):
        print('hello {}'.format(cls))
        
```

```python
a=YourClass()
```

```python
a.instance_say_hello()
```

```python
a.class_say_hello()
```
